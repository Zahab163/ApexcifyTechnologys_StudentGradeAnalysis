## Analyze Student's Grade

Analyze student's grades:
*  load a csv file containing student names and score in 3 subjects
* Calculate each students average scores
* Find the student with highest average score
* tools pandas ,python
* Sample column Name| Maths | English |Science
* Key Concept : mean() , max()

We are creating a simple .csv file frist

```python
import pandas as pd

#  Create data as a dictionary
data = {
    "Name": ["Ali", "Sara", "Ahmed"],
    "Maths": [85, 90, 70],
    "English": [78, 88, 65],
    "Science": [92, 95, 80]
}


```

```python
#  Convert dictionary to DataFrame
df = pd.DataFrame(data)
```

```python
#  Save DataFrame to CSV
df.to_csv("students.csv", index=False)

print("CSV file created successfully!")
```

We convert it again in Dataframe to understand and visualize the data

```python
#  Load CSV file as DataFrame
df = pd.read_csv("students.csv")
```

```python
# Display DataFrame
df.head()
```

```python
#  Calculate average score for each student
df["Average"] = df[["Maths", "English", "Science"]].mean(axis=1)
```

```python
#  Find student with highest average
highest_avg_student = df.loc[df["Average"].idxmax()]
```

```python
# See the Results
print("All students with their averages:")
print(df)
```

```python
# Check the student with highest average score
print("\nStudent with highest average score:")
print(highest_avg_student["Name"], "->", highest_avg_student["Average"])
```

In this exercise, we create our own CSV file, load it into pandas, calculate each student’s average score, and identify the top performer using simple functions like mean() and max(). With just a few lines of Python, we turned raw data into meaningful insights—showing how powerful and easy data analysis can be.
